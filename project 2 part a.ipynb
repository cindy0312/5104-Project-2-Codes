# ***Heart Disease Dataset***

```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd
import numpy as np

heart_disease = fetch_ucirepo(id=45) 

X = heart_disease.data.features 
y = heart_disease.data.targets 

print(heart_disease.variables) 
print(X.columns)
print(y.columns)

df = pd.concat([X, y], axis=1)
df = df.dropna()
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
print(y.value_counts().sort_index())

y_binary = y.copy()
y_binary = (y_binary > 0).astype(int)
print(y_binary.value_counts().sort_index())

heart_disease_binary = pd.concat([X, y_binary], axis=1)

heart_disease_binary.to_csv("heart_disease_binary.csv", index=False)
```

```python
import matplotlib.pyplot as plt

numerical = {"age", "trestbps", "chol", "thalach", "oldpeak", "ca"}
categorical = {"sex", "cp", "fbs", "restecg", "exang", "slope", "thal"}

for col in numerical:
    plt.figure(figsize=(6,4))
    plt.hist(X[col].dropna())
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show() 

for col in categorical:
    counts = X[col].value_counts().sort_index()
    plt.bar(counts.index.astype(str), counts.values, color="gray")
    plt.title(f"Bar Chart of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show() 

counts = y["num"].value_counts().sort_index()
plt.bar(counts.index.astype(str), counts.values, color="gray")
plt.title(f"Bar Chart of {"num"}")
plt.xlabel("num")
plt.ylabel("Count")
plt.show() 
```
